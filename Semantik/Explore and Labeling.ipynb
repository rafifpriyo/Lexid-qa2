```python
import json

sample_q6 = pd.read_csv('Lexid/questions/SampleQ6.csv')

with open('semantic/dataset/annotations.json') as f:
    d = json.load(f)
    # print(d)
    annotations = d['annotations']

all_question = []
all_ner = []

for i in range(len(annotations[1:151])):
  question = annotations[i+1][0]
  entities = annotations[i+1][1]['entities']
  ner = []
  for ner_label in entities:
    ner.extend([ner_label[2] for w in range(len(question[ner_label[0]:ner_label[1]].split(' ')))])
  question = question.strip('\r"')
  if len(question.split()) != len(ner):
    print(i)
    print('not same')
    print(question)
    print(ner)
  all_question.append(question)
  all_ner.append(ner)

df = pd.DataFrame(data={'question': all_question, 'answer': sample_q6['A'], 'ner': all_ner, 'complete_sentence': sample_q6['Q'], 'rule': sample_q6['reg_id']})
df['template_index'] = 'TS'
df.to_csv('semantic/dataset/sampleQ6ner.csv', index=False)
```
